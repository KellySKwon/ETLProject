```python
import pandas as pd
import numpy as np

#censuslink url https://factfinder.census.gov/faces/tableservices/jsf/pages/productview.xhtml?src=bkmk
#state url https://www.factmonster.com/us/postal-information/state-abbreviations-and-state-postal-codes
```

```python
url = 'https://www.numbeo.com/quality-of-life/region_rankings_current.jsp?region=021'
```

```python
#scrape and read as pandas df
tables = pd.read_html(url)
tables
```

```python
quality_df = tables[2]
len(quality_df['Rank'])
```

```python
# populate rank starting with 1
quality_df['Rank'] = [x for x in range(1,len(quality_df)+1)]
```

```python
#show clean rank
quality_df.head()
```

```python
#Find country and populate
quality_df['Country'] = ''
quality_df.loc[quality_df['City'].str.contains('Canada', case=False), 'Country'] = 'Canada'
quality_df.loc[quality_df['City'].str.contains('United States', case=False), 'Country'] = 'United States'
quality_df
```

```python
#Limit only where Country = United States
quality_US = quality_df[quality_df['Country'] == 'United States'].copy()
quality_US
```

```python
#Split City, State, Country by Comma
quality_City = quality_US.copy()
quality_City[['Metro','State','Country']] = quality_City.City.str.split(', ',expand=True).copy()
quality_City
```

```python
#Re-Rank City - find # of row
len(quality_City['Metro'])
```

```python
# populate rank for Only Cities in US starting with 1
quality_City['Rank'] = [x for x in range(1,len(quality_City)+1)]
quality_City
```

```python
#Scrape state acroynym table
stateurl='https://www.factmonster.com/us/postal-information/state-abbreviations-and-state-postal-codes'
#scrape and read as pandas df
scrapeState = pd.read_html(stateurl,header=0)
```

```python
#convert to df object
state_df = scrapeState[0]
state_df.head()
```

```python
#Rename state_df columns to set join with quality_City
state_name_df = state_df.rename(columns={'State':'State_Long','Postalcode':'State'})
state_name_df.tail(10)
```

```python
quality_City['State'].unique()
```

```python
state_name_df['State'].unique()
```

```python
state_name_df['State_Long'].unique()
```

```python
# Join the long State Name to Quality City 
quality_Index = quality_City.merge(state_name_df, on='State', how='left')
quality_Index
```

```python
#check columns
quality_Index.columns
```

```python
#Rename Columns to Export into MySQL
quality_final = quality_Index.rename(columns={'Quality of Life Index':'QOL_Index',
                             'Purchasing Power Index':'PP_Index',
                             'Safety Index':'Safety_Index',
                             'Health Care Index':'Health_Index',
                             'Cost of Living Index': 'CL_Index',
                             'Property Price to Income Ratio':'PPIR_Index',
                             'Traffic Commute Time Index':'TCT_Index',
                             'Pollution Index':'Pollution_Index',
                             'Climate Index':'Climate_Index'})
```
